```python
import sklearn
from sklearn import tree, datasets, metrics
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
import graphviz
```

```python
df = pd.read_csv(r'/home/max/data.csv', names=['PC', 'Target', 'Branch Outcome'])

df
```

```python
import csv

results = []
with open("/home/max/data.csv") as csvf:
    reader = csv.reader(csvf, delimiter=',')
    for row in reader:
        results.append(row)
```

```python
data = [x[:2] for x in results]
target = [x[2:] for x in results]
```

```python
data_1 = [x[:1] for x in results]
data_2 = [x[1:2] for x in results]
target_2 = [x[2:] for x in results]

data_1 = [[int(x) for x in sublist] for sublist in data_1]
data_2 = [[int(x) for x in sublist] for sublist in data_2]
target_2 = [[int(x) for x in sublist] for sublist in target]

data_1, data_2, target_2
```

```python
classification_tree = tree.DecisionTreeClassifier()

clf_tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=10)

classification_tree = classification_tree.fit(data, target)

clf_tree = clf_tree.fit(data, target)

dot_data = tree.export_graphviz(classification_tree, out_file=None,  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph.render("Gini") 

dot_data_2 = tree.export_graphviz(clf_tree, out_file=None,  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph_2 = graphviz.Source(dot_data_2)  
graph_2.render("Entropy") 

graph
```

```python
graph_2
```

```python
train_data, test_data, train_labels, test_labels = sklearn.model_selection.train_test_split(data, target,
                                                                   test_size=None, random_state=3)

clf = tree.DecisionTreeClassifier(random_state=1)
clf.fit(train_data, train_labels)

predictions = clf.predict(test_data)
metrics.accuracy_score(test_labels, predictions)
```
